```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Sample time series data
time = np.array([1, 2, 3, 4, 5, 6])
values = np.array([10, 15, 13, 20, 18, 25])

# Create a cubic spline model
cs = CubicSpline(time, values)

# Generate smooth time steps
time_smooth = np.linspace(1, 6, 100)
values_smooth = cs(time_smooth)

# Plot results
plt.scatter(time, values, color='red', label='Original Data')
plt.plot(time_smooth, values_smooth, label='Cubic Spline Fit')
plt.legend()
plt.show()
```
